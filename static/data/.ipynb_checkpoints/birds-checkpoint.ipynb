```python
import pandas as pd
file = 'HBW-BirdLife_Checklist_Version_3.xlsx'
```

```python
data = pd.read_excel(file)
data.columns = data.iloc[0]
data = data.drop(0)
```

```python
import openpyxl

wb = openpyxl.load_workbook(file)
ws = wb.active
```

```python
ws.cell(row=10000, column=5).hyperlink.target.split()
```

```python
hyperlinks = []
for row in range(3,ws.max_row+1):
    hyperlinks.append(ws.cell(row=row, column=5).hyperlink.target)
```

```python
data['hyperlink'] = hyperlinks
```

```python
names = [
    'Order','Family name','Family','CommonName','ScientificName',
    'Authority','TaxonomicTreatment', 'IUCNRedListCategory','Synonyms', 
    'AlternativeCommonNames', 'TaxonomicNotes','TaxonomicSource', 'SISRecID','SpcRecID', 'Hyperlink'
]
data.columns = names
data.head()
```

```python
data.to_csv("Birds.csv", index=False)
data.tail()
```
